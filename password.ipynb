Password Generator


```python
import random
import string

def generate_password(length=12, include_uppercase=True, include_digits=True, include_special=True):

  lowercase_letters = string.ascii_lowercase
  uppercase_letters = string.ascii_uppercase
  digits = string.digits
  special_symbols = string.punctuation

  characters = lowercase_letters
  if include_uppercase:
    characters += uppercase_letters
  if include_digits:
    characters += digits
  if include_special:
    characters += special_symbols

  password = ''.join(random.choice(characters) for _ in range(length))
  return password

if __name__ == "__main__":
  length = int(input("Enter the length of password: "))
  include_uppercase = input("Include uppercase letters? (yes/no): ").lower() == 'yes'
  include_digits = input("Include digits? (yes/no): ").lower() == 'yes'
  include_special = input("Input sprcial symbols? (yes/no): ").lower() == 'yes'

  password = generate_password(length, include_uppercase, include_digits, include_special)
  print(f"Generated Password: {password}")
```
